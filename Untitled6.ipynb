```python
import numpy as np
import pandas as pd
import plotly_express as px
import matplotlib.pyplot as plt
import streamlit as st
```

```python
q = float(input("enter value of flowrate in Stb/day = "))
k = float(input("enter value of permeability in md"))
Bo = float(input("enter value of formation volume factor in bbl/Stb = "))
viscosity = float(input("enter value of oil viscosity in cp = "))
h = float(input("enter value of resrvoir thickness in ft = "))
```

```python
Ct = float(input("enter value of total compressibility in 1/psi = "))
porosity = float(input("enter value of rock porosity in friction = "))
rw = float(input("enter value of wellbore radius in ft = "))
df = pd.read_excel("Testingdata.xlsx")
```

```python
df
```

```python
df["delp"] = 4412 - df["flowing bottome hole pressure(psi)"]
```

```python
df
```

```python
df["delT"] = np.zeros(len(df))
```

```python
df
```

```python
df["delT"][1:] = np.diff(df["t(hr)"])
```

```python
df
```

```python
df["d(delP)"] = np.zeros(len(df))
df["d(delP)"][1:] = np.diff(df["delp"])
```

```python
df
```

```python
df["P\'"] = df["t(hr)"] *(df["d(delP)"]/df["delT"])
```

```python
df["PD"] = 0.00708*k*h*df["delp"]/(q*Bo*viscosity)
```

```python
df
```

```python

df["td"] = 0.000264*k*df["t(hr)"]/(porosity*viscosity*Ct*rw**2)
```

```python
df
```

```python
df["del td"] = np.zeros(len(df)) 
df["del PD"] = np.zeros(len(df))
```

```python
df["del td"][1:] = np.diff(df["td"])
df["del PD"][1:] = np.diff(df["PD"])
df["PD\'"] = df["td"]*(df["del PD"]/df["del td"])
```

```python
df
```

```python

plt.style.use('default')
plt.figure(figsize = (15,7))
plt.loglog(df["td"],df["PD"] , c = "red",label = "PD v/s tD")
plt.loglog(df["td"],df["PD\'"] , c = "b",label = "PD\' v/s tD")
plt.loglog(df["t(hr)"],df["delp"] , c = "r", label = "delp v/s t",marker = "x")
plt.loglog(df["t(hr)"],df["P\'"] , c = "b",label = "P\' v/s t",marker = 'o')
plt.axhline(y = 0.5, c = "k", label = "y = 0.5", lw = 5)
plt.axhline(y = 38, c = "brown", label = "y = 38", lw = 5)
plt.title("Type Curves in Well Test Analysis.",fontsize=40)
plt.ylim(0,1500)
plt.ylabel("delta P , P\' , PD , PD\'", fontsize=20)
plt.xlabel("t, tD", fontsize=20)

plt.axvspan(0.001,1000,color='yellow',alpha=0.5,label='Actual Data-Plots')
plt.axvspan(10000,100000000,color='orange',alpha=0.5,label='Type-Curve Data-Plots')

plt.legend()
plt.grid()
```
