## Decline curve anylish

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

## Exponential Decline 

```python
t_1m = 1 #month 
q_1m = 96 #stb/day

t_0m = 0 #month
q_0m = 100 #stb/day

b = np.log(q_0m/q_1m)/(t_1m - t_0m)
```

```python
b
```

```python
# rate of end of one year 
q_12m = q_0m * np.exp(-b *12)
```

```python
q_12m
```

```python
t = np.arange(0,100,2)
t.sort()
```

```python
t
```

```python
q = q_0m * np.exp( -b * t )
Np = (q_0m - q) * 365/(b *12)
Np
```

```python
plt.plot(t,q , c = "r" ,  label= " t v/s q" , linestyle = ':' , marker = "x")
plt.plot(t,Np/500 , c = "g" , label = "t v/s Np/500", linestyle = ":",  marker ="o")
style.use('ggplot')
plt.legend()
plt.xlabel("time")
plt.ylabel("Production rate")
plt.grid(True, color = 'b')
 
```

## Semilog plot

```python
ax = plt.subplot()
ax.semilogy(t,q , c = "r" ,  label= " t v/s q" , linestyle = ':' , marker = "x")
ax.set_xlabel("time(month)")
ax.set_ylabel(" log(Flow rate (STB/day)) ")
style.use('ggplot')
plt.legend()
plt.grid(True, color = 'b')
ax2 = ax.twinx()
ax2.plot(t,Np , c = "g" , label = "t v/s Np", linestyle = ":",  marker ="o")

ax2.set_ylabel("total production (STB) ")
plt.legend()
```

## Harmonic Decline

```python
t_1m = 1 #month 
q_1m = 96 #stb/day

t_0m = 0 #month
q_0m = 100 #stb/day

b = (q_0m/q_1m - 1)/(t_1m - t_0m)
```

```python
b
```

```python
t = np.arange(0.1,100,0.5)
t.sort()
```

```python
q = q_0m / ( 1 + b *t)
Np = (np.log(q_0m) - np.log(q)) *((q_0m *365) /(12*b)) 
q
Np
```

```python
q
```

```python
ax = plt.subplot()
ax.plot(t,q , c = "r" ,  label= " t v/s q" , linestyle = ':' , marker = "x")
ax.set_xlabel("time(month)")
ax.set_ylabel("Flow rate (STB/day) ")
style.use('ggplot')
plt.legend()
plt.grid(True, color = 'b')
ax2 = ax.twinx()
ax2.plot(t,Np , c = "g" , label = "t v/s Np", linestyle = ":",  marker ="o")

ax2.set_ylabel("total production (STB) ")
plt.legend()
```

## semilog plot total production v/s rate 

```python
plt.semilogy(Np,q, c ="b", label =" Semilog plot Np v/s rate")
style.use('ggplot')
plt.legend()
plt.grid(True, color = 'b')
plt.xlabel("total production")
plt.ylabel("production rate")
```
